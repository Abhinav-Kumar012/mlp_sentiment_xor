```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xor_df = pd.read_csv('datasets/xor.csv',index_col=0)
xor_df
```

```python
fig, ax = plt.subplots(figsize=(12,10))
sc = ax.scatter(x=xor_df['X1'], y=xor_df['X2'], c=xor_df['label'], cmap='coolwarm')
fig.colorbar(sc, ax=ax, label='value', orientation='vertical')
plt.show()
```
